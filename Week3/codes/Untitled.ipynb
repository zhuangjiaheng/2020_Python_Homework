```python
from pyecharts import options as opts
from pyecharts.charts import Pie, Timeline
from pyecharts.faker import Faker

attr = Faker.choose()
tl = Timeline()
for i in range(2015, 2020):
    pie = (
        Pie()
        .add(
            "商家A",
            [list(z) for z in zip(attr, Faker.values())],
            rosetype="radius",
            radius=["30%", "55%"],
        )
        .set_global_opts(title_opts=opts.TitleOpts("某商店{}年营业额".format(i)))
    )
    tl.add(pie, "{}年".format(i))
tl.render("timeline_pie.html")
```

```python
import random
import pandas as pd
import math

cats0 = ['驯鹿','飞机','火车','船','汽车']
color0 = ['#D6A5DD','#F7DC73','#EBBDBF','#76CBE8','#A4DFE5']
year0 = list(range(1977,2021))
year = []
for i in year0:
    for j in range(len(cats0)):
        year.append(i)

cats = cats0*len(year0)
color = color0*len(year0)
value = []
for i in year:
    value.append(math.pow(random.randint(0,50), 2))#求平方，拉开数据
df = pd.DataFrame()
df['year'] = year
df['value'] = value
df['cats'] = cats
df['color'] = color
df['csum']=df.groupby(['cats'])['value'].cumsum()
```

```python
df[df['year']==1999]
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Timeline

tl = Timeline()
for i in range(1977,2019):
    df_sub = df[df['year']==i].sort_values(by='csum')
    cats_list = list(df_sub['cats'])
    csum_list = list(df_sub['csum'])
    color_list = list(df_sub['color'])

    bar = (
        Bar()
        .add_xaxis(cats_list)
        .add_yaxis("", csum_list,label_opts=opts.LabelOpts(position="right",font_weight="bold"),category_gap=30)
        .reversal_axis()
        .set_global_opts(
            title_opts=opts.TitleOpts("跑得快 (时间: {} 年)".format(i),pos_left=350,padding=[30,20]),
            xaxis_opts=opts.AxisOpts(max_=50000)
        )
    )
    tl.add(bar, "{}年".format(i))
    tl.add_schema(play_interval=300, is_loop_play=True)#播放间隔，是否循环播放
tl.render("timeline_bar_reversal_dynamic_data.html")
```
