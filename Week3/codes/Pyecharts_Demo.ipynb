### Demo1

```python
from pyecharts import options as opts
from pyecharts.charts import Map, Timeline
from pyecharts.faker import Faker

tl = Timeline()
for i in range(2015, 2020):
    map0 = (
        Map()
        .add("商家A", [list(z) for z in zip(Faker.provinces, Faker.values())], "china")
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Map-{}年某些数据".format(i)),
            visualmap_opts=opts.VisualMapOpts(max_=200),
        )
    )
    tl.add(map0, "{}年".format(i))
    tl.add_schema(play_interval=300, is_loop_play=True)#播放间隔，是否循环播放
tl.render_notebook()
    # tl.render("timeline_map.html")
```

### Demo2

```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Timeline
from pyecharts.faker import Faker

x = Faker.choose()
tl = Timeline()

for i in range(2015, 2020):
    print(i,Faker.values())
    print(i,Faker.values())
    bar = (
        Bar()
        .add_xaxis(x)
        .add_yaxis("angry", Faker.values())
        .add_yaxis("disguist", Faker.values())
        .add_yaxis("fear", Faker.values())
        .add_yaxis("joy", Faker.values())
        .add_yaxis("sadness", Faker.values())
        .set_global_opts(title_opts=opts.TitleOpts("某商店{}年营业额".format(i)))
    )
    tl.add(bar, "{}年".format(i))
    tl.add_schema(play_interval=500, is_loop_play=True)#播放间隔，是否循环播放
tl.render_notebook()
```
